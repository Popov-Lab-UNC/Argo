```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import IPythonConsole

# Configure RDKit settings
IPythonConsole.ipython_useSVG = True
rdDepictor.SetPreferCoordGen(True)

# Get the molecules from the response
mols_data = json.loads(response_body['molecules'])

# Convert SMILES to RDKit molecules
mols = [Chem.MolFromSmiles(m['sample']) for m in mols_data]

# Draw the molecules in a grid
img = Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(300,300), 
                          legends=[f"Score: {m['score']:.3f}" for m in mols_data])
img
```
